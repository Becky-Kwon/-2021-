```python
## 개인 주피터에서 실행할 시
# 데이터 로딩
buy_2018_1 = pd.read_csv('./기업데이터/buy2018_1.csv')
buy_2018_2 = pd.read_csv('./기업데이터/buy2018_2.csv')
buy_2019_1 = pd.read_csv('./기업데이터/buy2019_1.csv')
buy_2019_2 = pd.read_csv('./기업데이터/buy2019_2.csv')
sns_2018_1 = pd.read_csv('./기업데이터/sns2018_1.csv')
sns_2018_2 = pd.read_csv('./기업데이터/sns2018_2.csv')
sns_2019_1 = pd.read_csv('./기업데이터/sns2019_1.csv')
sns_2019_2 = pd.read_csv('./기업데이터/sns2019_2.csv')
```

```python

# 기존의 1부터 시작하는 이미 부여된 index 열 삭제
buy_2018_1 = buy_2018_1.drop('Unnamed: 0', axis=1)
buy_2018_2 = buy_2018_2.drop('Unnamed: 0', axis=1)
buy_2019_1 = buy_2019_1.drop('Unnamed: 0', axis=1)
buy_2019_2 = buy_2019_2.drop('Unnamed: 0', axis=1)
sns_2018_1 = sns_2018_1.drop('Unnamed: 0', axis=1)
sns_2018_2 = sns_2018_2.drop('Unnamed: 0', axis=1)
sns_2019_1 = sns_2019_1.drop('Unnamed: 0', axis=1)
sns_2019_2 = sns_2019_2.drop('Unnamed: 0', axis=1)

# 이후의 데이터 merge와 concat을 고려해 칼럼명 통일
buy_2018_1.columns = ['date','gender','age','big_cat','sm_cat','quantity']
buy_2018_2.columns = ['date','gender','age','big_cat','sm_cat','quantity']
buy_2019_1.columns = ['date','gender','age','big_cat','sm_cat','quantity']
buy_2019_2.columns = ['date','gender','age','big_cat','sm_cat','quantity']
sns_2018_1.columns = ['date','big_cat','sm_cat','count']
sns_2018_2.columns = ['date','big_cat','sm_cat','count']
sns_2019_1.columns = ['date','big_cat','sm_cat','count']
sns_2019_2.columns = ['date','big_cat','sm_cat','count']

# 데이터 합치기
# 연도별 구분 데이터셋
buy_2018 = pd.concat([buy_2018_1,buy_2018_2], ignore_index=True)      # 2018년의 구매데이터
buy_2019 = pd.concat([buy_2019_1,buy_2019_2], ignore_index=True)      # 2019년의 구매데이터
sns_2018 = pd.concat([sns_2018_1,sns_2018_2], ignore_index=True)      # 2018년의 SNS 노출 데이터
sns_2019 = pd.concat([sns_2019_1,sns_2019_2], ignore_index=True)      # 2019년의 SNS 노출 데이터
# 연도 구분 없는 총 데이터셋
buy = pd.concat([buy_2018,buy_2019], ignore_index=True)
sns = pd.concat([sns_2018,sns_2019], ignore_index=True)
# 뷰티, 식품, 냉난방가전 분류에 따라 분류된 buy, sns 총 데이터셋
buy_beauty = buy[buy['big_cat']== '뷰티'].reset_index(drop=True)
buy_food = buy[buy['big_cat']== '식품'].reset_index(drop=True)
buy_appliance = buy[buy['big_cat']== '냉난방가전'].reset_index(drop=True)
sns_beauty = sns[sns['big_cat']== '뷰티'].reset_index(drop=True)
sns_food = sns[sns['big_cat']== '식품'].reset_index(drop=True)
sns_appliance = sns[sns['big_cat']== '냉난방가전'].reset_index(drop=True)
```
